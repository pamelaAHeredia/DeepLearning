## Sigmoide

Realice 30 ejecuciones independientes de la configuración seleccionada para respaldar sus afirmaciones referidas a la precisión obtenida tanto para el conjunto de entrenamiento como para el de testeo. Utilice un máximo de 200 iteraciones.

```python
def ejercicio_9(funcion):
  iteraciones = 200
  cota = 1e-03

  alfas = [0.1, 0.2, 0.3]

  for a in alfas:
    modelo_s = NeuronaGradiente(alpha=a, n_iter=iteraciones, cotaE=cota, FUN=funcion)

    accuracy_train = 0
    accuracy_test = 0

    max_test = -1
    max_train = -1

    it_max_test = -1
    it_max_train = -1

    veces_train = 0
    veces_test = 0

    total_it = 0

    for i in range(30):
      # separo
      X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.4)

      # normalizo ejemplos
      X_train = preprocessing.StandardScaler().fit_transform(X_train)
      X_test = preprocessing.StandardScaler().fit_transform(X_test)

      # Esto porque las etiquetas me quedan como -1 y 1. Qué hago acá??
      if (funcion == 'tanh'):
        y_test = y_test.apply(lambda x: 1 if x == 1 else -1)
        y_train = y_train.apply(lambda x: 1 if x == 1 else -1)

      # Entreno
      modelo_s.fit(X_train, y_train)

      # Pruebo
      acc_train = modelo_s.accuracy(X_train, y_train)
      acc_test = modelo_s.accuracy(X_test, y_test)

      # Calculo y coso
      if acc_train > max_train:
        max_train = acc_train
        it_max_train = i

      if (acc_test > max_test):
        max_test = acc_test
        it_max_test = i

      if acc_train == 1:
        veces_train += 1
      if acc_test == 1:
        veces_test +=1
      accuracy_train += acc_train
      accuracy_test += acc_test
      total_it += modelo_s.i_

      # print(f'Salida: {modelo_s.predict(X_test)}')

    print(f'Alfa: {a}')
    print(f'accuracy Test: {accuracy_test/30} \nAccuracy Train: {accuracy_train/30} \nIteraciones: {total_it/30} ')
    print(f'Veces (datos test): {veces_test} \nVeces (datos train): {veces_train}')
    print(f'Ite accuracy: \n test:{it_max_test} \n train: {it_max_train}')
    print('------------------------------------------------------------')
```
